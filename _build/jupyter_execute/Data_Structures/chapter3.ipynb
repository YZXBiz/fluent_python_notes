# Standard API of Mapping Types


The collections.abc module provides the Mapping and MutableMapping ABCs describing the interfaces of dict and similar types.

```{note}
The main value of the ABCs is documenting and formalizing the standard interfaces for mappings, and serving as criteria for `isinstance test` in code that needs to support mappings in a broad sense.
```


````python
import collections.abc as abc

my_dict = {}
isinstance(my_dict, abc.Mapping)
````

````python
isinstance(my_dict, abc.MutableMapping)
````

```{tip}
Using isinstance with an ABC is often better than checking whether a funciton argument is of the conrete dict type, because then alternative mapping types can be used.
```

UserDict class and all concrete mapping classes in the standard library encapsulate the basic dict in their implementation, which in turn is built on a hash table.
Therefore, they all share the limitation that the keys must be hashable (the values need not be hashable, only the keys). If you need a refresher, the next section explains.


# What is Hashable


```{note}
An object is hashable if it has a hash code which never changes during its lifetime (it needs a `__hash__` method), and can be compared to other objects (it needs an `__eq__` method).

Hashable objects which compare equal must have the same hash value.
```


- Numeric types and flat immutable types (str and bytes) are hashable.
- Container types are hashable if they are immutable and all contained objects are also hashable.
- A frozenset is always hashable, because its elements must be hashable by definition.
- A tuple is hashable only if all its items are hashable.

```{python}
t1 = (1, 2, (3, 4))
t2 = (1, 2, [3, 4])
hash(t1) # 2528502973977326415
hash(t2) # TypeError: unhashable type: 'list'
```
